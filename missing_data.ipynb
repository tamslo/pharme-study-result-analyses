## Missing Data


```python
import pandas as pd

from modules.definitions.constants import SCORE_COLUMN
from modules.definitions.types import Survey
from modules.study_progress import get_study_progress
from modules.survey_results.get_data import get_survey_results

progress_message, potentially_lost_cases, partial_surveys = get_study_progress()

print("Partial surveys:")  # noqa: T201
for time_point_name, partial_time_point_surveys in partial_surveys.items():
    if len(partial_time_point_surveys.index) == 0:
        continue
    log_message = f"... at {time_point_name}:"
    print(log_message)  # noqa: T201
    display(partial_time_point_surveys)

print("Missing single answers:")  # noqa: T201
missing_data_counts = []
for survey in Survey:
    if not survey.value.is_final:
        continue
    survey_data = get_survey_results(survey)
    if SCORE_COLUMN in survey_data.columns:
        survey_data = survey_data.drop(SCORE_COLUMN, axis=1)
    missing_data = survey_data[survey_data.isna().any(axis=1)]
    if len(missing_data) == 0:
        continue
    for column in missing_data.columns:
        missing_answers_count = missing_data[column].isna().sum()
        if missing_answers_count == 0:
            continue
        missing_data_counts.append(
            [survey.value.name, column, missing_answers_count],
        )
display(
    pd.DataFrame(
        missing_data_counts,
        columns=["survey", "question", "na_count"],
    ).set_index(["survey", "question"]),
)
```
